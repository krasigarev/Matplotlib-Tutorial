```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import mplcyberpunk
```

```python
fig = plt.figure(figsize=(8,5))
ax: plt.Axes = fig.gca()

x = np.linspace(0, 5 * np.pi, 200)
y_1 = 1.5 * np.cos(x)
y_2 = 2.5 * np.sin(x)

ax.plot(x, y_1, label="y_1")
ax.plot(x, y_2, label="y_2")

ax.grid()
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(6,10), sharex=True, sharey=True)
ax: plt.Axes = fig.gca()

x = np.linspace(0, 5 * np.pi, 200)
y_1 = 1.5 * np.cos(x - 1)
y_2 = 2.5 * np.cos(x + 1)

p1 = axs[0].plot(x, y_1, 'tab:blue')
p2 = axs[1].plot(x, y_2, 'tab:orange')

for i, ax in enumerate(axs):
    ax.grid()
    ax.set_ylabel('f(x)')
axs[1].set_xlabel('x')
axs[1].legend(handles = [p1[0], p2[0]], labels= ['y1', 'y2'])

plt.show()
```

```python
rng = np.random.default_rng(seed=0)

fig = plt.figure(figsize=(8,6))

x = np.linspace(0, 10, 100)

for index in range(1, 3 * 2 + 1):
    ax = fig.add_subplot(3, 2, index)
    params = rng.uniform(0.5, 1, 3)
    b = params[0] * np.sin(params[1] * x) + params[2]
    ax.plot(x, b)
    ax.tick_params(axis='both', which='both', length=0)
    plt.setp(ax.get_xlabel(), visible=False)
    plt.setp(ax.get_ylabel(), visible=False)

plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca()

x = np.linspace(0, 10, 25)
y = 0.5 * x * np.sin(x) + 0.8 * x

rng = np.random.default_rng(seed=0)
y_err = rng.random(y.shape)
y_err[y_err < 0.5] = 0.5

ax.errorbar(x, y, yerr=y_err, ecolor='r')

plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca()

x = np.linspace(0, 10, 25)
y = 0.5 * x * np.sin(x) + 0.8 * x

rng = np.random.default_rng(seed=0)
y_err = rng.random(y.shape)
y_err[y_err < 0.5] = 0.5

ax.fill_between(x, y - y_err, y + y_err)
ax.plot(x, y, color='tab:orange')

plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs_list: list[plt.Axes] = list(axs.flatten())

rng = np.random.default_rng(seed=0)
N = 4

m = rng.uniform(-1, 1, N)
s = rng.uniform(1, 2, N)
v = rng.normal(m, s, size=(100, N))

axs_list[0].boxplot(v)
axs_list[1].boxplot(v, orientation='horizontal')

plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 14))
axs_list: list[plt.Axes] = list(axs.flatten())

rng = np.random.default_rng(seed=0)
N = 4

m = rng.uniform(-1, 1, N)
s = rng.uniform(1, 2, N)
v = rng.normal(m, s, size=(100, N))

axs_list[0].violinplot(v, showmedians=True)
axs_list[1].violinplot(v, showmedians=True, orientation='horizontal')

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.gca()

x = np.linspace(0, 10, 100)
y = -0.5 * x * np.sin(x) + 0.5 * x

index_y_max = np.argmax(y)
y_value = y[index_y_max]
x_value = x[index_y_max]

ax.plot(x, y , color="tab:orange")

ax.text(x=8.6, y=0,
        s="Text can\n be put\n anywhare!")
ax.annotate(text="Maximum", 
            color='blue',
            xy=(x_value, y_value),
            xytext=(6, 7),
            arrowprops=dict(facecolor="b")
            )
plt.show()
```
